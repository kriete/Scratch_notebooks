```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

#### The txt file looks like this:

![csv_file](image.png)

### Read this into a pandas DataFrame:

```python
year = 2011
df = pd.read_csv('01_May_' + str(year) +'.txt', header=None)
```

#### Merging of the two header lines

```python
a = df.iloc[0:2,:].fillna('').values
b = np.column_stack([a[0], a[1]])
d = []
for c in b:
    d.append(' '.join(c).strip())
```

#### Setting the header columns

```python
df = df.iloc[2::]
df.columns = d
```

#### Date conversion (for convenience)

```python
df.Date = df.Date.str[:] + ' ' + str(year)
df.index = pd.to_datetime(df['Date'])
```

#### Plotting all columns except non-numeric ones

```python
for c in df.columns[3::]:
    df[c] = df[c].astype('float')
    plt.figure(figsize=(8, 5))
    df[c].plot(grid=True, lw=2)
    plt.ylabel(c)
```
